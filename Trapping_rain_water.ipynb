1. Take minimum of `heights[0]` and `heights[-1]`.
2. Check if it is `>` with `max(heights[1:-1])`.
3. If True, split `heights` into `a` and `b`, and return `f(a) + f(b)`.
4. Else calculate water in given list `heights` by :
>   1. Taking the $min(first, last)$ i.e $M = min(heights[0], heights[-1]).$
>   2. Minus all values between first and last by $M.$
>   3. Increment result by these values.
>   4. Return result.
5. Returning sum of all the results.

###  4. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
import numpy as np

def trapped_water(heights):
  if len(heights) < 3:
    return 0

  Min = min(heights[0], heights[-1])
  index = np.argmax(heights[1:-1]) + 1

  if Min < heights[index]:
    return trapped_water(heights[:index+1]) + trapped_water(heights[index:])

  result = 0

  for i in range(1, len(heights)-1):
    result += Min - heights[i]

  return result
```

```python
def trapped_water2(heights):
  water = 0

  for i in range(len(heights)):
    lmax = max(heights[:i+1])
    rmax = max(heights[i:])
    water += min(lmax, rmax) - heights[i]

  return water
```

### 7. Come up with a correct solution for the problem. State it in plain English.

Come with the optimized correct solution and explain it in simple words below:

1. Make 2 pointers for lmax and rmax.
2. Through Binary Search find lmax for right water and rmax for left water.
3. Change lmax and rmax accordingly.
4. Else, calculate water and change lo or hi as:
>    1. If $l[lo] < lmax ($where $l[hi] > (l[lo], lmax)): water += lmax - l[lo]$ and increment $lo.$
>    2. If $l[hi] > rmax ($where $l[hi] <= (l[lo], rmax)): water += rmax - l[hi]$ and decrement $hi.$
5. Return Water.
6. End.

### 8. Implement the solution and test it using example inputs. Fix bugs, if any.

```python
def Trap_rainwater(heights):
  lmax, rmax = 0, 0
  lo, hi = 0, len(heights)-1
  water = 0

  while lo < hi:
    if heights[lo] < heights[hi]:
      if heights[lo] < lmax:      water += lmax - heights[lo]
      else:                       lmax = heights[lo]
      lo += 1

    else:
      if heights[hi] < rmax:      water += rmax - heights[hi]
      else:                       rmax = heights[hi]
      hi -= 1
  return water
```
